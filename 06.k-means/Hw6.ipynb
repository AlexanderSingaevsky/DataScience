```python
import warnings
import matplotlib.pyplot as plt
import pandas as pd
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
warnings.filterwarnings("ignore")
```

```python
data_2d = pd.read_csv('data/data_2d.csv', names=["_n", "_x", "_y"])
print(data_2d.shape)
```

```python
mnist = pd.read_csv('data/mnist.csv', names=[f'x_{i}' for i in range(786)])
mnist.fillna(0, inplace=True)
print(mnist.shape)
```

```python
k_elbow_data_2d = KElbowVisualizer(KMeans(), k=(1,10)).fit(data_2d) 
```

```python
k_elbow_mnist = KElbowVisualizer(KMeans(), k=(1,10)).fit(mnist) 
```

```python
#data_2d
kmeans_data_2d = KMeans(n_clusters=2, n_init = 'auto').fit(data_2d)
plt.figure(figsize=[5,5])
plt.scatter(data_2d['_x'], data_2d['_y'], cmap='bwr', c=kmeans_data_2d.labels_)
plt.show()
```

```python
pca_mnist = pd.DataFrame(PCA(n_components=2).fit_transform(mnist))
kmeans_mnist = KMeans(n_clusters=5, n_init = 'auto').fit(pca_mnist)
plt.figure(figsize=[5, 5])
plt.scatter(pca_mnist[0], pca_mnist[1], c=kmeans_mnist.labels_, cmap='brg')
plt.show()
```
